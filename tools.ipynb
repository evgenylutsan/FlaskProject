```python
import fitz
import os

t = os.path.basename(pdf_path)
title = os.path.splitext(t.rpartition('_')[-1])[0]
doc = fitz.open(pdf_path)
page = doc.load_page(0)
pix = page.get_pixmap()
output = "C:\\Users\lutsa\\OneDrive\\Рабочий стол\\python_project\\prewiew\\prewiew - " + title + ".png"
pix.save(output)

path_pic = output
```

```python
# pict_path = r"C:\Users\lutsa\OneDrive\Рабочий стол\python_project\prewiew\prewiew.png"
def pict_binary(pict_path):
    f = open(pict_path, 'rb')
    pic_bin = f.read()
    return pic_bin
bin_pic = pict_binary(pict_path = path_pic)
print(bin_pic)
```
